Cargar librerías y datos

Primero cargamos las librerías que vamos a usar. Pandas para manejar los datos, Seaborn y Matplotlib para los gráficos.  
Después abrimos nuestro archivo titanic.csv y echamos un vistazo a las primeras filas para asegurarnos de que se ha cargado correctamente.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar CSV
titanic = pd.read_csv('./titanic.csv')
titanic.head()  # Verificacion
```

Distribución de edades

Aquí hacemos un histograma para ver cómo se distribuyen las edades de los pasajeros del Titanic.  
El color azul claro lo usamos solo para que quede más bonito, y ponemos etiquetas en los ejes y un título para que sea fácil de entender.

```python
plt.figure(figsize=(8,5))
sns.histplot(titanic['Age'], bins=30, color='skyblue')
plt.xlabel('Edad')
plt.ylabel('Número de pasajeros')
plt.title('Distribución de edades')
plt.savefig('histograma_edades.png')
plt.show()
```

Precios de los billetes

Vamos a ver los 5 billetes más caros y los 5 más baratos, y los representamos en un gráfico de dispersión.
Así podemos ver la edad de los pasajeros con los billetes extremos y calcular la media del precio de esos billetes.

```python
caros = titanic.nlargest(5, 'Fare')
baratos = titanic.nsmallest(5, 'Fare')
seleccion = pd.concat([caros, baratos])

plt.figure(figsize=(8,5))
sns.scatterplot(data=seleccion, x='Fare', y='Age', hue='Fare', palette='cool', s=100)
plt.xlabel('Precio del billete')
plt.ylabel('Edad')
plt.title('5 billetes más caros y 5 más baratos')
plt.savefig('scatter_billetes.png')
plt.show()

print("Media del precio de estos billetes:", seleccion['Fare'].mean())
```

Supervivientes por sexo

Aquí hacemos un gráfico que nos muestra el porcentaje de hombres y mujeres que sobrevivieron.  
Usamos colores distintos para los supervivientes y los que no, y ponemos etiquetas para que se entienda fácilmente.

```python
plt.figure(figsize=(6,4))
sns.histplot(data=titanic, x='Sex', hue='Survived', multiple='fill', palette='Set2')
plt.ylabel('Porcentaje de supervivientes')
plt.title('% de supervivientes por sexo')
plt.savefig('supervivientes_sexo.png')
plt.show()
```

Relaciones entre variables

Por último, hacemos un pairplot para ver cómo se relacionan algunas variables importantes:  
Edad (age)  
Precio del billete (fare)  
Clase del pasajero (pclass)  
Supervivencia (survived)  

El pairplot nos ayuda a ver posibles patrones entre las variables.

```python
cols = ['Age', 'Fare', 'Pclass', 'Survived']
sns.pairplot(titanic[cols], hue='Survived', palette='coolwarm', diag_kind='kde')
plt.savefig('pairplot.png')
plt.show()
```

Conclusión

Hemos visto que la edad de los pasajeros está bastante distribuida, que los billetes más caros suelen ir con pasajeros de edad media/alta, y que la supervivencia varía mucho según el sexo.  

En general, los gráficos ayudan mucho a visualizar los datos y entender patrones rápidamente.
